```python
import matplotlib.pyplot as plt
import numpy as np

# Synthetic heart rate data
time = np.arange(0, 24, 0.5)  # Hours
heart_rate = np.random.normal(loc=72, scale=5, size=len(time))  # Beats per minute

# Plotting the heart rate trend
plt.figure(figsize=(10, 6))
plt.plot(time, heart_rate, label="Heart Rate", color='blue')
plt.axhline(y=100, color='red', linestyle='--', label="High Heart Rate Threshold")
plt.axhline(y=60, color='green', linestyle='--', label="Low Heart Rate Threshold")
plt.title("Heart Rate Trends Over 24 Hours")
plt.xlabel("Time (hours)")
plt.ylabel("Heart Rate (BPM)")
plt.legend()
plt.grid()
plt.show()

```
